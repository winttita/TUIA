## Extraer el mes de nacimiento de cada persona en una nueva columna. ¿Cuál fue el mes con mayor cantidad de nacimientos?

```python
datos_personas['Mes_nacimiento'] = datos_personas['Fecha_nacimiento_cod'].dt.month
datos_personas.head()
```

```python
# Diccionario meses
meses = {
    1: 'enero', 2: 'febrero', 3: 'marzo', 4: 'abril',
    5: 'mayo', 6: 'junio', 7: 'julio', 8: 'agosto',
    9: 'septiembre', 10: 'octubre', 11: 'noviembre', 12: 'diciembre'
}

datos_personas['Mes_nacimiento_nombre'] = datos_personas['Mes_nacimiento'].map(meses)
datos_personas.head()
```

```python
datos_personas['Mes_nacimiento_nombre'].value_counts()
```
